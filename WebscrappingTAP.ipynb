```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd 
```

```python
url = "https://www.flytap.com/en-pt/destinations"

user_agent = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36"}

response = requests.get(url, headers=user_agent)
response.status_code
```

```python
soup = BeautifulSoup(response.content)
soup
```

```python
destinations_country = soup.find_all("span", attrs= {"class":"country"})

len(destinations_country)

destinations_country

country = [text.text for text in destinations_country]
country
```

```python
destinations_city = soup.find_all("span", attrs= {"class":"city"})

len(destinations_city)

destinations_city

city = [text.text for text in destinations_city]
city
# len(city)
```

```python
destinations_local= soup.find_all("div", attrs= {"class":"local"})

len(destinations_local)

destinations_local 

for text in destinations_local:
    print(text.text)
    
# country + city 
# gives me also categories for each [country city] combination 
#  ex.: 

# <span class="country">Morocco</span>
#  <span class="city">Agadir</span>
#  </div>,
#  <div class="local categories">
#  <span class="icn-svg svg-icn-destinos-beach"><span class="prop-hide--sreader">Beach</span></span>
#  <span class="icn-svg svg-icn-destinos-gastronomy"><span class="prop-hide--sreader">Gastronomy</span></span>
#  <span class="icn-svg svg-icn-destinos-night"><span class="prop-hide--sreader">Nightlife</span></span>
```
